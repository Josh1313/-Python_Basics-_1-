# Dataframes de Pandas

Un DataFrame de Python es una tabla de 2-dimensiones, potencialmente hetereogéna y con ejes etiquetados (filas y columnas).

Propiedades de los DataFrames:
  - Columnas pueden ser de distintos tipos de datos
  - modificable de tamaño
  - ejes etiquetados (filas y columnas)
  - se pueden realizar operaciones aritméticas en filas y columnas

Los DataFrames de Pandas consisten básicamente de 3 elementos principales:
-  datos (*data*)
-  filas (*rows*)
-  columnas (*columns*)

## Pivot Table

```python
# Tabla pivote para calcular el promedio de la columna 'C' para cada combinación única de 'A' y 'B'
pivot_table1 = pd.pivot_table(df, values='C', index=['A', 'B'], aggfunc=np.mean)

print(pivot_table1)
```

```python
# Tabla pivote para calcular la suma de la columna 'D' para cada valor único en la columna 'A'
pivot_table2 = pd.pivot_table(df, values='D', index='A', aggfunc=np.sum)

print(pivot_table2)
```

```python
# Tabla pivote para calcular el máximo y el mínimo de la columna 'C' para cada valor único en la columna 'B'
pivot_table3 = pd.pivot_table(df, values='C', index='B', aggfunc=[np.max, np.min])

print(pivot_table3)
```

Generar un DataFrame con los datos del fichero. Para importar usaremos Pandas.

```python
print("Dimensiones del DataFrame:", df.shape)
print("Número de datos en el DataFrame:", df.size)
print("Nombres de las columnas:", df.columns.tolist())
print("Nombres de las filas (índice):", df.index.tolist())
print("Tipos de datos de las columnas:")
print(df.dtypes)
print("Las 10 primeras filas:")
print(df.head(10))
print("Las 10 últimas filas:")
print(df.tail(10))
```

```python
print("Datos del pasajero con identificador 148:")
print(df.loc[147])
```

```python
# Paso 4: Mostrar los nombres de las personas que iban en primera clase ordenadas alfabéticamente
print("Nombres de las personas que iban en primera clase ordenados alfabéticamente:")
print(df[df['Pclass'] == 1]['Name'].sort_values())
```

```python
# Paso 5: Mostrar el porcentaje de personas que sobrevivieron y murieron
sobrevivieron = df[df['Survived'] == 1].shape[0]
murieron = df[df['Survived'] == 0].shape[0]
total_pasajeros = df.shape[0]
porcentaje_sobrevivieron = (sobrevivieron / total_pasajeros) * 100
porcentaje_murieron = (murieron / total_pasajeros) * 100

print("Porcentaje de personas que sobrevivieron: {:.2f}%".format(porcentaje_sobrevivieron))
print("Porcentaje de personas que murieron: {:.2f}%".format(porcentaje_murieron))
```

```python
# Paso 6: Mostrar el porcentaje de personas que sobrevivieron en cada clase
porcentaje_sobrevivieron_clase = df.groupby('Pclass')['Survived'].mean() * 100
print("Porcentaje de personas que sobrevivieron en cada clase:")
print(porcentaje_sobrevivieron_clase)
```

```python
# Paso 7: Eliminar del DataFrame los pasajeros con edad desconocida
df = df.dropna(subset=['Age'])
```

```python
# Paso 8: Mostrar la edad media de las mujeres que viajaban en cada clase
edad_media_mujeres_clase = df[df['Sex'] == 'female'].groupby('Pclass')['Age'].mean()
print("Edad media de las mujeres que viajaban en cada clase:")
print(edad_media_mujeres_clase)
```

```python
# Paso 9: Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no
df['MenorEdad'] = df['Age'] < 18
```

```python
# Paso 10: Mostrar el porcentaje de menores y mayores de edad que sobrevivieron en cada clase
porcentaje_sobrevivieron_menores_clase = df[df['MenorEdad'] == True].groupby('Pclass')['Survived'].mean() * 100
porcentaje_sobrevivieron_mayores_clase = df[df['MenorEdad'] == False].groupby('Pclass')['Survived'].mean() * 100

print("Porcentaje de menores de edad que sobrevivieron en cada clase:")
print(porcentaje_sobrevivieron_menores_clase)

print("Porcentaje de mayores de edad que sobrevivieron en cada clase:")
print(porcentaje_sobrevivieron_mayores_clase)
```
